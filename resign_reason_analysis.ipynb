```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_excel('improving.xlsx')
df.head()
```

```python
# convert TanggalHiring to datetime and create tahun_hiring column from the year 
df['TanggalHiring'] = pd.to_datetime(df['TanggalHiring'])
```

```python
df_tahun_hiring = df['TanggalHiring'].dt.year
```

```python
df['TanggalResign'] = pd.to_datetime(df['TanggalResign'],errors='coerce')
```

```python
df_tahun_resign = df['TanggalResign'].dt.year
```

```python
# outer join df_tahun_hiring and df_tahun_resign
df_tahun = pd.concat([df_tahun_hiring,df_tahun_resign],axis=1)
```
